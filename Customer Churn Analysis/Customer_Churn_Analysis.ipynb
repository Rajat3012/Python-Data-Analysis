```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Customer Churn.csv")
```

replacing blanks with 0 as tenure is 0 and no total charges are recorded


```python
df["TotalCharges"]=df['TotalCharges'].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
#checking duplicate value for particular column
df["customerID"].duplicated().sum()
```

```python
def conv(value):
  if value==1:
    return 'yes'
  else:
    return 'no'
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df
```

```python
plt.title("Number of customer churned out")
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.title("Percentage of Churned Customer")
gb=df.groupby("Churn").agg({"Churn":"count"})

plt.pie(gb["Churn"],labels=gb.index,autopct="%1.2f%%")

plt.show()
```

```python
plt.title('Gender wise churn')
sns.countplot(x="gender",data=df,hue="Churn")
plt.show()
```

```python
plt.title("Churn by Senior Citizen")
sns.countplot(x="SeniorCitizen",data=df,hue="Churn")
plt.show()
```

```python
sns.histplot(x="tenure",data=df,bins=72,hue="Churn")
plt.show()
```

people who have used our services for a long time have stayed and people who have used our services for one or two months churned

```python

plt.figure(figsize=(4,4))
plt.title("Count of customers by contract")
ax=sns.countplot(x="Contract",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

People who have month-to-month contract are likely to churn more as compare to 1 year or 2 year contract

```python

# List of columns to create count plots
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
n_cols = 3  # Number of columns in the grid
n_rows = -(-len(columns) // n_cols)  # Calculate rows needed using ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()  # Flatten the axes array for easy indexing

# Plot each count plot
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i],hue=df["Churn"])  # Create count plot
    axes[i].set_title(f"Count Plot of {column}")  # Add title
    axes[i].set_ylabel("Count")  # Label y-axis
    axes[i].set_xlabel(column)  # Label x-axis

# Remove any empty subplots if columns do not fill the grid
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
