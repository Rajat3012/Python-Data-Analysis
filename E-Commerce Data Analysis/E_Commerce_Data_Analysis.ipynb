```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Sample - Superstore.csv",encoding='latin1')
```

```python
#Convert data type of dates from object to datetime format
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df.head(5)
```

```python
#extracting month,year and order day
df["Order Month"]=df["Order Date"].dt.month
df["Order Year"]=df["Order Date"].dt.year
df["Order Day"]=df["Order Date"].dt.day
```

```python
df
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

**Monthly Sales Analysis**

```python
sales_by_month=df.groupby("Order Month")["Sales"].sum().reset_index()
fig=px.bar(sales_by_month,x="Order Month",y="Sales",title="Monthly Sales Analysis")
fig.show()
```

**Sales by Category**

```python
sales_by_category=df.groupby("Category")["Sales"].sum().reset_index()
fig=px.pie(sales_by_category,values="Sales",names="Category",hole=0.5,color_discrete_sequence=colors.qualitative.Pastel)
fig.update_traces(textposition="inside",textinfo="percent+label")
fig.update_layout(title="Sales by Category")
fig.show()
```

**Sales based on sub categories**

```python
sales_by_sub=df.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
sales_by_sub
```

```python
fig=px.bar(sales_by_sub,x="Sub-Category",y="Sales",title="Sales by Sub-Category")
fig.show()
```

**Monthly Profit from sales**

```python
profit_from_sales=df.groupby("Order Month")["Profit"].sum().reset_index()
fig=px.line(profit_from_sales,x="Order Month",y="Profit",title="Monthly Profit from Sales")
fig.show()
```

**Profit by Category**

```python
profit_by_category=df.groupby("Category")["Profit"].sum().reset_index()
profit_by_category=df.groupby("Category")["Profit"].sum().reset_index()
fig=px.pie(profit_by_category,values="Profit",names="Category",hole=0.5,color_discrete_sequence=colors.qualitative.Pastel)
fig.update_traces(textposition="inside",textinfo="percent+label")
fig.update_layout(title="Profit by Category")
fig.show()
```

**Profit by Sub-Category**

```python
profit_by_sub=df.groupby("Sub-Category")["Profit"].sum().reset_index()
fig=px.bar(profit_by_sub,x="Sub-Category",y="Profit",title="Profit by Sub-Category")
fig.show()
```

**Sales and Profit by Customer Segment**

```python
sales_profit_by_segment=df.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
color_palette=colors.qualitative.Pastel
fig=go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment["Segment"],y=sales_profit_by_segment["Sales"],name="Sales",marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment["Segment"],y=sales_profit_by_segment["Profit"],name="Profit",marker_color=color_palette[1]))

fig.update_layout(title="Sales and Profit by Customer Segment",xaxis_title="Customer Segment",yaxis_title="Amount")

fig.show()
```

**Sales to Profit Ratio**

```python
sales_profit_by_segment=df.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
sales_profit_by_segment["Sales to Profit Ratio"]=sales_profit_by_segment["Sales"]/sales_profit_by_segment["Profit"]
sales_profit_by_segment
```
